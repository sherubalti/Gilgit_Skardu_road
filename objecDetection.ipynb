```python
import ipywidgets as widgets
from IPython.display import display
from ipywidgets import FileUpload, Output
import matplotlib.pyplot as plt
import numpy as np
import cv2  # OpenCV for image manipulation

# Create a FileUpload widget for image selection
file_upload = FileUpload(accept='.jpg, .jpeg, .png', multiple=False)

# Create an Output widget for displaying results
output_widget = Output()

# Function to perform road sign detection
def detect_road_sign(change):
    with output_widget:
        output_widget.clear_output()
        for uploaded_filename in file_upload.value:
            content = file_upload.value[uploaded_filename]['content']
            # Process the uploaded image here and perform detection
            # You can use OpenCV or your trained YOLOv8 model for this
            # Display the detected result (e.g., bounding boxes) on the image
            img = cv2.imdecode(np.frombuffer(content, np.uint8), -1)
            # Perform detection and draw bounding boxes here
            plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
            plt.axis('off')
            plt.show()

# Attach the detection function to the file upload button
file_upload.observe(detect_road_sign, names='value')

# Display the widgets
display(file_upload, output_widget)
```
